### Number Based Patterns

```python
# 1st for loop,variable i => i corresponds to the row
# 2nd for loop,variable j => j corresponds to the columns
```

```python
#### a) 
n = int(input('Enter a number: '))  # 5
for i in range(1,n+1):     # i = 1,2,3,4,5
    for j in range(1,n+1): # j = 1,2,3,4,5
        print(i,end=" ")
    print()
```

```python
#### b) 
n = int(input('Enter a number: '))  # 5
for i in range(1,n+1):     # i = 1,2,3,4,5
    for j in range(1,n+1): # j = 1,2,3,4,5
        print(j,end=" ")#adding space between 2 elements
    print()#entering next line after each row is completed
```

```python
#### c) 
n = int(input('Enter a number: '))  # 5
for i in range(n,0,-1):     
    for j in range(1,n+1): 
        print(i,end=" ")
    print()
```

```python
#### d) 
n = int(input('Enter a number: '))  # 5
for i in range(1,n+1):    
    for j in range(n,0,-1):
        print(j,end=" ")
    print()
```

```python
#### e) 
n = int(input('Enter a number: '))  # 5
for i in range(1,n+1):     # i = 1,2,3,4,5
    for j in range(1,n+1): # j = 1,2,3,4,5
        if i==j:
            print(i,end=" ")
        else:
            print(0,end=" ")
    print()
```

```python
#### f)
n = int(input('Enter a number: '))  # 5
for i in range(1,n+1):     # i = 1,2,3,4,5
    for j in range(1,n+1): # j = 1,2,3,4,5
        if i+j==n+1:
            print(i,end=" ")
        else:
            print(0,end=" ")
    print()
```

```python
#### g)
n = int(input('Enter a number: '))  # 5
for i in range(n,0,-1):     
    for j in range(n,0,-1): 
        if i+j==n+1:
            print(i,end=" ")
        else:
            print(0,end=" ")
    print()
```

```python
#### h)
n = int(input('Enter a number: '))  # 5
for i in range(1,n+1):     
    for j in range(n,0,-1): 
        if i+j==n+1:
            print(j,end=" ")
        else:
            print(0,end=" ")
    print()
```

```python
#### i) 
n = int(input('Enter a number: '))  # 5
for i in range(1,n+1):     # i = 1,2,3,4,5
    for j in range(1,n+1): # j = 1,2,3,4,5
        if i==j or i+j==n+1:
            print(i,end=" ")
        else:
            print(0,end=" ")
    print()
```

```python
#### j) 
n = int(input('Enter a number: '))  # 5
for i in range(n,0,-1):     # i = 5,4,3,2,1
    for j in range(n,0,-1): # j = 5,4,3,2,1
        if i==j or i+j==n+1:
            print(i,end=" ")
        else:
            print(0,end=" ")
    print()
```

```python
#### k) 
n = int(input('Enter a number: '))  # 5
for i in range(1,n+1):     # i = 1,2,3,4,5
    for j in range(1,n+1): # j = 1,2,3,4,5
        if (i+j)%2 == 0 :
            print("1",end=" ")
        else:
            print("0",end=" ")
    print()
```

```python
#### l
n = int(input('Enter a number: '))
for i in range(1,n+1):
    for j in range(1,n+1):
        print(i+j-1,end=' ')
    print()
```

### Alphabet Series

```python
# ASCII code
# A-Z = 65 - 90
# a-z = 97-122
```

```python
print(chr(65),chr(90))
print(ord('A'),ord('Z'))
print(chr(97),chr(122))
print(ord('a'),ord('z'))
```

### Star Shaped Patterns

```python
#### a)
n = int(input('Enter a number: '))  # 5
for i in range(1,n+1):     # i = 1,2,3,4,5
    for j in range(1,n+1): # j = 1,2,3,4,5
        print('*',end=" ")
    print()
```

```python
#### b)
n = int(input('Enter a number: '))  # 5
for i in range(1,n+1):     # i = 1,2,3,4,5
    print('*'*i)
```

```python
#### c)
n = int(input('Enter a number: '))  # 5
for i in range(n,0,-1):     
    print('*'*i)
```

```python
#### d)
n = int(input('Enter a number: '))  #5
for i in range(1,n+1):
    for j in range(1,n+1):
        if i+j>n:
            print('*',end="")
        else:
            print(' ',end="")
    print()
```

```python
#### e)
n = int(input('Enter a number: '))
for i in range(n,0,-1):
    for j in range(1,n+1):
        if i+j>n:
            print("*",end="")
        else:
            print(" ",end="")
    print()
```

```python
#### f)
n = int(input('Enter a number: '))
for i in range(1,n+1):
    print(' '*(n-i),'* '*i)
```

```python
#### g)
n = int(input('Enter a number: '))
for i in range(n,0,-1):
    print(' '*(n-i),'* '*i)
```

```python
#### h)
n = int(input('Enter a number: '))
for i in range(1,n):
    print(' '*(n-i),'* '*i)
for i in range(n,0,-1):
    print(' '*(n-i),'* '*i)
```

```python
#### i)
n=int(input('Enter a number: '))
for i in range(1,n):
    print('*'*i)
for i in range(n,0,-1):
    print('*'*i)
```

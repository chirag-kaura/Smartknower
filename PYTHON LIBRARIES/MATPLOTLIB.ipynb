```python
import matplotlib.pyplot as plt ## Viz library
import numpy as np
```

### Different Types of Plots

#### 1) Bar Chart
Used to compare different categories of data

```python
a = [1,2,3,4,5,6,7,8]
b = [14,18,9,13,19,16,15,17]
print(len(a),len(b))
```

```python
# a is represented along x-axis, b is represented along y-axis
print(a)
print(b)
plt.bar(a,b)
plt.show()
```

```python
plt.bar(a,b,label='b',color=['orange'])
plt.title('a vs b')
plt.xlabel('values of a')
plt.ylabel('values of b')
plt.xlim(0,14)
plt.ylim(0,25)
plt.legend()
plt.grid()
plt.savefig('fig1.png')
plt.show()
```

```python
print(a)
print(b)
print(max(b),min(b))
print(a[b.index(max(b))],a[b.index(min(b))])
```

```python
plt.bar(a,b,label='b',color='orange')
plt.bar(a[b.index(max(b))],max(b),label='highest',color='blue')
plt.bar(a[b.index(min(b))],min(b),label='lowest',color='red')
plt.title('a vs b')
plt.xlabel('values of a')
plt.ylabel('values of b')
plt.legend()
plt.grid()
plt.show()
```

#### 2) Line Chart
Used to represent continuos data wrt. date time vlaues.

```python
plt.plot(a,b,label='b',color='blue',marker='o')
plt.title('a vs b')
plt.xlabel('values of a')
plt.ylabel('values of b')
plt.legend()
plt.grid()
plt.show()
```

```python
s1 = np.random.randint(1,20,10)
s2 = np.random.randint(1,20,10)
s3 = np.random.randint(1,20,10)
print(len(s1))
```

```python
# lw= linewidth
plt.plot(list(range(10)),s1,label='s1-sales',color='blue',marker='o',lw=2,linestyle=":")
plt.plot(list(range(10)),s2,label='s2-sales',color='orange',marker='^',lw=1,linestyle="-.")
plt.plot(list(range(10)),s3,label='s3-sales',color='maroon',marker='X',lw=3,linestyle="--")
plt.title('s1, s2 and s3')
plt.xlabel('Day')
plt.ylabel('Sales')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.bar(a,b,label='b_bar',color='maroon')
plt.plot(a,b,label='b',color='blue',marker='o')
plt.title('a vs b')
plt.xlabel('values of a')
plt.ylabel('values of b')
plt.legend(loc=10)
plt.grid()
plt.show()
```

#### 3) Scatter Chart
Used to represent distribution of 2 continuous varaibles

```python
plt.scatter(s1,s2,label='s1 vs s2',s=120,color='red',marker='d')
plt.xlabel('s1')
plt.ylabel('s2')
plt.title('s1 vs s2')
plt.legend()
plt.grid()
```

#### 4) Histogram
Used to represent frequency distribution

```python
from collections import Counter
```

```python
x = [3,5,4,6,7,5,2,1,2,3,4,3,1,2,3,2,5,2,4,5,5,7,5,6,5,7,4,3,2,1]
print(x)
```

```python
r = Counter(x)
print(r)
```

```python
plt.hist(x,edgecolor='black')
plt.show()
```

```python
print(r)
plt.hist(x,edgecolor='black',bins=[1,2,3,4,5,6,7],color='orange')
plt.grid()
plt.show()
# 1st bin => [1,2) - 1 inclusive, 2 exclusive => 1:3
# 2nd bin => [2,3) - 2 inclusive, 3 exclusive => 2:6
# 3rd bin => [3,4) - 3 inclusive, 4 exclusive => 3:5
# 4th bin => [4,5) - 4 inclusive, 5 exclusive => 4:4
# 5th bin => [5,6) - 5 inclusive, 6 exclusive => 5:7
# 6th bin => [6,7] - 6 inclusive, 7 exclusive => 6:2,7:3 => total =5
```

```python
print(r)
plt.hist(x,edgecolor='black',bins=[1,2,3,4,5,6,7,8],color='tan')
plt.grid()
plt.show()
# 1st bin => [1,2) - 1 inclusive, 2 exclusive => 1:3
# 2nd bin => [2,3) - 2 inclusive, 3 exclusive => 2:6
# 3rd bin => [3,4) - 3 inclusive, 4 exclusive => 3:5
# 4th bin => [4,5) - 4 inclusive, 5 exclusive => 4:4
# 5th bin => [5,6) - 5 inclusive, 6 exclusive => 5:7
# 6th bin => [6,7] - 6 inclusive, 7 exclusive => 6:2
# 7th bin => [7,8] - 7 inclusive, 8 exclusive => 7:3,8:0 => total =3
```

#### 5) Boxplot
1) Used to represent percentile distribution.<br>
2) It returns 5 point summary<br>

a) Q1 = 25 percentile<br>
b) Q2 = 50 percentile (Median)<br>
c) Q3 = 75 percentile<br>
d) Min = Q1 - 1.5 * IQR<br>
e) Max = Q3 + 1.5 * IQR<br>

IQR (Inter Quartile Range)<br>
Outliers = >max or <min

```python
x = [18,15,17,16,20,3,14,16,13,17,18,10,11,12,13,17,19,20,45,14]
print(len(x))
```

```python
plt.boxplot(x)
plt.grid()
plt.show()
```

#### 6) Pie Chart
Used to represent percentage distribution

```python
sub = ['Python','DS','ML','DL','Tableau']
scores = [78,84,85,76,92]
per_python = (scores[0]/sum(scores))*100
print(per_python)
```

```python
plt.pie(x=scores,labels=sub,autopct='%.2f%%',explode= (0,0,0,0,0.1))
plt.show()
```
